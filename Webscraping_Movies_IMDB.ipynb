```python
#importing all packages required
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from IPython.display import clear_output
```

```python
#creating empty lists inorder to append
name=[]
year=[]
certificate=[]
rating=[]
duration=[]
genre=[]
votes=[]
metascore=[]

#base url of the site
base_url='https://www.imdb.com/search/title/?release_date=2021-01-01,2021-12-31&sort=num_votes,desc&start='

#scraping  movie info 
for i in range(1,3502,50):
  time.sleep(3)
  print(f'Scraping for {i} movies')
  clear_output(wait=True)
    
  url=base_url+str(i)+'&ref_=adv_nxt'
  req=requests.get(url)
  soup=BeautifulSoup(req.text,'html')
    
    
  #movie_name
  n=soup.findAll('h3',attrs={'class':'lister-item-header'})
  for j in n:
    name.append(j.a.text)
        
    
  #year_of_release
  y=soup.findAll('h3',{'class':'lister-item-header'})
  for k in y:
    year.append((k.find('span',attrs={'class':'lister-item-year text-muted unbold'}).text.replace('(','').replace(')','')))
        
    
  #duration
  d=soup.findAll('div',{'class':'lister-item mode-advanced'})
  for l in d:
    du=l.find('span',{'class':'runtime'})
    if du is not None:
      duration.append(du.text)
    else:
      duration.append("Not available")
        
      
  #certification
  c=soup.findAll('div',{'class':'lister-item mode-advanced'})
  for m in c:
    certificate.append(m.find('p',{'class':'text-muted'}).span.text)
        
  #ratings
  rate=soup.findAll('div',attrs={'class':'ratings-bar'})
  for n in rate:
    rating.append(n.strong.text)
        
  #metascore
  meta=soup.findAll('div',{'class':'ratings-bar'})
  for o in meta:
    m=o.find('span',attrs={'class':'metascore'})
    if m is not None:
      metascore.append(m.text.strip())
    else:
      metascore.append('Not available')
            
  #genre
  g=soup.findAll('div',{'class':'lister-item mode-advanced'})
  for l in g:
   a=l.find('span',{'class':'genre'})
   if a is not None:
     genre.append(a.text.strip())
   else:
     genre.append("Not available")
  
        
        
  #votes
  v=soup.findAll('p',{'class':'sort-num_votes-visible'})
  for q in v:
    votes.append(q.find('span',{'name':'nv'}).text)
```

```python
print(len(name))
print(len(year))
print(len(votes))
print(len(genre))
print(len(metascore))
print(len(duration))
print(len(certificate))
print(len(rating))
```

```python
#merging all datas into a single frame and final data set will be displayed
data= pd.DataFrame({'Movie':name,
                   'Released Year':year,
                   'Certificate':certificate,
                   'Duration':duration,
                   'Genre':genre,
                   'Rating':rating,
                   'Metascore':metascore,
                   'Votes':votes})
data
```

```python
data.to_csv('movie details.csv')
```
